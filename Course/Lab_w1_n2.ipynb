```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
# x_train is the input variable (size in 1000 square feet)
# y_train is the output variable (price in 1000s of dollars)

x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
# m is number of training examples
print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0]
print(f"Number of training examples(m) = {m}")
```

```python
i = 0 # change this to 1 to see the second training example (x^1, y^1)

x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")	
```

```python
# Plot the data (training) points
plt.scatter(x_train, y_train, marker='x', color='r')
#Set the title
plt.title('Housing Prices')
#Set the y-axis label
plt.ylabel('Price in 1000s of dollars')
#Set the x-axis label
plt.xlabel('Size in 1000 square feet')
plt.show()
```

```python
w = 200
b = 100
print(f"w : {w}")
print(f"b : {b}")
```

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of linear model
    Args:
        x (ndarray(m,)) : Data, m examples
        w,b (scalar)    : model parameters
    Returns:
        y (ndarray(m,)) : target values
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w*x[i] + b
    
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b)

#Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b', label='Our Prediction')

#Plot the data (training) points
plt.scatter(x_train, y_train, marker='x', color='r', label='Actual Values')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
